```python
wine_2015 = ecx[ecx['Item'] == 'Wine']['Y2015'].sum()
wine_2018 = ecx[ecx['Item'] == 'Wine']['Y2018'].sum()

wine_2015, wine_2018
```

```python
production_2014 = ecx[ecx['Element'] == 'Production']['Y2014'].sum()
production_2014
```

```python
# Calculate the total number of unique countries
unique_countries = ecx['Area'].nunique()
unique_countries
```

```python
# Calculate the sum of each element in 2018
sum_2018 = ecx.groupby('Element')['Y2018'].sum()

# Find the element with the highest sum
highest_sum_element = sum_2018.idxmax()

highest_sum_element
```

```python
# Calculate the total number of missing data in 2016
missing_data_2016 = ecx['Y2016'].isnull().sum()

# Calculate the total number of data points in 2016
total_data_2016 = ecx['Y2016'].shape[0]

# Calculate the percentage of missing data in 2016
percentage_missing_2016 = (missing_data_2016 / total_data_2016) * 100

missing_data_2016, round(percentage_missing_2016, 3)
```

```python
# Calculate the total Import Quantity in Algeria in 2018
import_quantity_algeria_2018 = ecx[(ecx['Area'] == 'Algeria') & (ecx['Element'] == 'Import Quantity')]['Y2018'].sum()

import_quantity_algeria_2018
```

```python
# Calculate the correlation between 'Element Code' and each year
year_correlation = ecx[['Element Code', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].corr()

year_correlation
```

```python
# Find the year with the highest correlation
max_corr_year = year_correlation.drop('Element Code', axis=1).idxmax().values[0]

max_corr_year
```
